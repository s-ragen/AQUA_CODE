```python
def plot_surface_v(dataset, case):
    b = np.linspace(-0.25, 0.25, 21, endpoint=True)
    t = np.linspace(-0.25, 0.25, 11, endpoint=True)
    
    fig = plt.figure(figsize=(14, 6))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
    plt.contourf(dataset['xh'], dataset['yq'], dataset['vo'].isel(zl=0).mean(dim='time'), 
                 b, cmap='RdBu_r', extend='both',  transform=ccrs.PlateCarree())
    cbar = plt.colorbar(ticks=t, boundaries=t, spacing='uniform', extend='both')
    cbar.ax.set_ylabel('[m s$^{-1}$]', fontsize=14)
    plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
    plt.title('{} Meridional Velocity'.format(case), fontsize=16)
    plt.show()

    b = np.linspace(-0.1, 0.1, 21, endpoint=True)
    fig = plt.figure(figsize=(14, 6))
    plt.contourf(dataset['xh'], dataset['yq'].sel(yq=slice(35, 75)),
                 dataset['vo'].sel(yq=slice(35, 75)).isel(zl=0).mean(dim='time'), b, cmap='RdBu_r', extend='both')
    cbar = plt.colorbar(ticks=t, boundaries=t, spacing='uniform', extend='both')
    cbar.ax.set_ylabel('[m s$^{-1}$]', fontsize=14)
    plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
    plt.title('{} Meridional Velocity'.format(case), fontsize=16)
    plt.show()
    return

def plot_int_v(dataset, case):
    dz = xr.zeros_like(dataset.zl) + 2.5
    dz[1:31] = dataset.zl.diff('zl').values

    Vz = (dataset.vo.isel(time=slice(-31,-1)).mean(dim='time')* dz).sum(dim='zl')
    # Vz = np.where(dataset.vo.isel(zl=0)==np.nan, Vz, np.nan)

    mask = np.isnan(dataset.vo.isel(zl=0).isel(time=-1))
    Vz_values = np.where(mask==False, Vz, np.nan)
    Vz.values = Vz_values

    b = np.linspace(-100, 100, 21, endpoint=True)
    t = np.linspace(-100, 100, 11, endpoint=True)

    fig = plt.figure(figsize=(14, 6))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
    plt.contourf(dataset['xh'], dataset['yq'], Vz, b, 
                 cmap='RdBu_r', extend='both',  transform=ccrs.PlateCarree())
    cbar = plt.colorbar(ticks=t, boundaries=t, spacing='uniform', extend='both')
    cbar.ax.set_ylabel('[m$^2$ s$^{-1}$]', fontsize=14)
    plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
    plt.title('{} Vertically Integrated Meridional Velocity'.format(case), fontsize=16)
    plt.show()

    fig = plt.figure(figsize=(14, 6))
    plt.contourf(dataset['xh'], dataset['yq'].sel(yq=slice(35, 75)),
                 Vz.sel(yq=slice(35, 75)), b, cmap='RdBu_r', extend='both')
    cbar = plt.colorbar(ticks=t, boundaries=t, spacing='uniform', extend='both')
    cbar.ax.set_ylabel('[m$^2$ s$^{-1}$]', fontsize=14)
    plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
    plt.title('{} Vertically Integrated Meridional Velocity'.format(case), fontsize=16)
    plt.show()
    return

def plot_int_v_diff(dataset1, dataset2, case1, case2):
    dz = xr.zeros_like(dataset1.zl) + 2.5
    dz[1:31] = dataset1.zl.diff('zl').values

    Vz1 = (dataset1.vo.isel(time=slice(-31,-1)).mean(dim='time')* dz).sum(dim='zl')
    # Vz = np.where(dataset.vo.isel(zl=0)==np.nan, Vz, np.nan)

    mask1 = np.isnan(dataset1.vo.isel(zl=0).isel(time=-1))
    Vz1_values = np.where(mask1==False, Vz1, np.nan)
    Vz1.values = Vz1_values

    Vz2 = (dataset2.vo.isel(time=slice(-31,-1)).mean(dim='time')* dz).sum(dim='zl')
    # Vz = np.where(dataset.vo.isel(zl=0)==np.nan, Vz, np.nan)

    mask2 = np.isnan(dataset2.vo.isel(zl=0).isel(time=-1))
    Vz2_values = np.where(mask2==False, Vz2, np.nan)
    Vz2.values = Vz2_values
    
    bounds = 50
    b = np.linspace(-bounds, bounds, 21, endpoint=True)
    t = np.linspace(-bounds, bounds, 11, endpoint=True)

    fig = plt.figure(figsize=(14, 6))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
    plt.contourf(dataset1['xh'], dataset1['yq'], Vz1-Vz2, b, 
                 cmap='RdBu_r', extend='both',  transform=ccrs.PlateCarree())
    cbar = plt.colorbar(ticks=t, boundaries=t, spacing='uniform', extend='both')
    cbar.ax.set_ylabel('[m$^2$ s$^{-1}$]', fontsize=14)
    plt.contour(dataset1['xh'], dataset1['yq'], mask1, [0.01], colors='gray', transform=ccrs.PlateCarree())
    plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
    plt.title('{}-{} Vertically Integrated Meridional Velocity'.format(case1, case2), fontsize=16)
    plt.show()

    fig = plt.figure(figsize=(14, 6))
    plt.contourf(dataset1['xh'], dataset1['yq'].sel(yq=slice(35, 75)),
                 Vz1.sel(yq=slice(35, 75))-Vz2.sel(yq=slice(35, 75)), b, cmap='RdBu_r', extend='both')
    cbar = plt.colorbar(ticks=t, boundaries=t, spacing='uniform', extend='both')
    cbar.ax.set_ylabel('[m$^2$ s$^{-1}$]', fontsize=14)
    plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
    plt.title('{}-{} Vertically Integrated Meridional Velocity'.format(case1, case2), fontsize=16)
    plt.contour(dataset1['xh'], dataset1['yq'], mask1, [0.01], colors='gray')
    plt.ylim([35,75])
    plt.show()
    return
# mask = np.zeros(sst.shape, dtype=bool)
#     mask = np.where(~np.isnan(sst), mask, 1)

#     mask_sss = np.ma.array(sst, mask=mask)
```
