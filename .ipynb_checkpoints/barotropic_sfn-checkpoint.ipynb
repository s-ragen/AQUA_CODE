```python
umo_slice = dz_straight.umo.isel(time=slice(70,100)).mean(dim='time')*1030
umo_sum = -umo_slice.sum(dim='z_l')
psi = umo_sum[:,::-1].cumsum(dim='xq')/1e6
psi.plot()
```

```python
uo_tmp = dy_straight.uo.isel(time=slice(70,100)).mean(dim='time')*dy_straight.zl
# u_bt = (uo_tmp/dy_straight.zl.sum(dim='zl')).sum(dim='zl')
u_bt = uo_tmp.sum(dim='zl')

fig, ax = plt.subplots(figsize=(10, 6))
u_bt.plot(ax=ax)

vo_tmp = dy_straight.vo.isel(time=slice(70,100)).mean(dim='time')*dy_straight.zl
# u_bt = (uo_tmp/dy_straight.zl.sum(dim='zl')).sum(dim='zl')
v_bt = vo_tmp.sum(dim='zl')

fig, ax = plt.subplots(figsize=(10, 6))
v_bt.plot(ax=ax)
```

```python
psi = -u_bt.cumsum(dim='yh')/1e6
psi.plot()
```

```python
# Psi = Int{y0:y}u(x,y) dy - Int{x_0:x}v(x,y_0) dx 

uo_tmp = dy_straight.uo.isel(time=slice(70,100)).mean(dim='time')*dy_straight.zl
# u_bt = (uo_tmp/dy_straight.zl.sum(dim='zl')).sum(dim='zl')
u_bt = uo_tmp.sum(dim='zl')

vo_tmp = dy_straight.vo.isel(time=slice(70,100)).mean(dim='time')*dy_straight.zl
# u_bt = (uo_tmp/dy_straight.zl.sum(dim='zl')).sum(dim='zl')
v_bt = vo_tmp.sum(dim='zl')

dy = np.diff(u_bt.yh)
dx = np.diff(v_bt.xh)

Psi = -grid_straight.cumsum(u_bt, 'y')*dy[0] + grid_straight.cumsum(v_bt, 'x')*dx[0]
(Psi/1e6).plot()
```

```python
mask = grid_straight.interp(dy_straight.vo.isel(zl=0).mean(dim='time'), 'x')
psi = Psi.where(mask==np.nan, np.nan)

(psi/1e6).plot()
```
