```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import warnings

#=================================================
# Modify paths to point to output files
#=================================================
# Case name
case = 'Both Coast'

# path to ocean_annual file
path_year = '/data/sragen/aquaplanet/MASTERS/both_coast/run/DATA/const_100yr/19000101.ocean_annual.nc'
dy = xr.open_dataset(path_year, decode_times=True)

# path to ocean_month file
path_month = '/data/sragen/aquaplanet/MASTERS/both_coast/run/DATA/const_100yr/19000101.ocean_month.nc'
dm = xr.open_dataset(path_month, decode_times=True)

# path to ice_month file
path_month = '/data/sragen/aquaplanet/MASTERS/both_coast/run/DATA/const_100yr/19000101.ice_month.nc'
di = xr.open_dataset(path_month, decode_times=True)

# path to core file with zonal wind stress
path_core = '/data2/sragen/CORE/CORE_new/u_10.idealized.nc'
dc = xr.open_dataset(path_core, decode_times=True)

# path to save figures
fig_path = 'const_100yr/FIGURES/'

#=================================================
# Ignore runtime warnings: mean of empty slice
#=================================================
warnings.filterwarnings("ignore", message="Mean of empty slice")
```

```python
di.sithick
di.SALTF
```

```python
sithick = di['sithick'][-31:,:,:]
sithick_djf = np.nanmean(sithick.where(di['time.season'] == 'DJF'), axis=0)
sithick_jja = np.nanmean(sithick.where(di['time.season'] == 'JJA'), axis=0)

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
plt.contourf(di['xT'], di['yT'], np.nanmean(sithick, axis=0), extend='both',  transform=ccrs.PlateCarree())
cbar = plt.colorbar(spacing='uniform', extend='both')
cbar.ax.set_ylabel('[m]', fontsize=14)
plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
plt.title('Yearly Mean Sea Ice Thickness', fontsize=16)
# plt.savefig(fig_path+'MLD_mean.png')
plt.show()

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
plt.contourf(di['xT'], di['yT'], sithick_djf, extend='both',  transform=ccrs.PlateCarree())
cbar = plt.colorbar(spacing='uniform', extend='both')
cbar.ax.set_ylabel('[m]', fontsize=14)
plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
plt.title('DJF Mean Sea Ice Thickness', fontsize=16)
# plt.savefig(fig_path+'MLD_DJF.png')
plt.show()

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
plt.contourf(di['xT'], di['yT'], sithick_jja, extend='both',  transform=ccrs.PlateCarree())
cbar = plt.colorbar(spacing='uniform', extend='both')
cbar.ax.set_ylabel('[m]', fontsize=14)
plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
plt.title('JJA Mean Sea Ice Thickness', fontsize=16)
# plt.savefig(fig_path+'MLD_JJA.png')
plt.show()
```

```python
SALTF = di['SALTF'][-31:,:,:]
SALTF_djf = np.nanmean(SALTF.where(di['time.season'] == 'DJF'), axis=0)
SALTF_jja = np.nanmean(SALTF.where(di['time.season'] == 'JJA'), axis=0)

b = np.linspace(-2.5e-8, 2.5e-8, 21, endpoint=True)
t = np.linspace(-2.5e-8, 2.5e-8, 11, endpoint=True)

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
plt.contourf(di['xT'], di['yT'], np.nanmean(SALTF, axis=0), 
             b,  extend='both', cmap='BrBG_r', transform=ccrs.PlateCarree())
cbar = plt.colorbar(ticks=t, spacing='uniform', extend='both')
cbar.ax.set_ylabel(r'$[kg\ m^{-2}\ S^{-1}]$', fontsize=14)
plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
plt.title('Yearly Mean Salt Flux from Ice', fontsize=16)
# plt.savefig(fig_path+'MLD_mean.png')
plt.show()

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
plt.contourf(di['xT'], di['yT'], SALTF_djf, b, cmap='BrBG_r',
             extend='both',  transform=ccrs.PlateCarree())
cbar = plt.colorbar(ticks=t, spacing='uniform', extend='both')
cbar.ax.set_ylabel(r'$[kg\ m^{-2}\ S^{-1}]$', fontsize=14)
plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
plt.title('DJF Mean Salt Flux from Ice', fontsize=16)
# plt.savefig(fig_path+'MLD_DJF.png')
plt.show()

fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180.0))
plt.contourf(di['xT'], di['yT'], SALTF_jja, b, cmap='BrBG_r',
             extend='both',  transform=ccrs.PlateCarree())
cbar = plt.colorbar(ticks=t, spacing='uniform', extend='both')
cbar.ax.set_ylabel(r'$[kg\ m^{-2}\ S^{-1}]$', fontsize=14)
plt.ylabel('Latitude', fontsize=14), plt.xlabel('Longitude', fontsize=14)
plt.title('JJA Mean Salt Flux from Ice', fontsize=16)
# plt.savefig(fig_path+'MLD_JJA.png')
plt.show()
```
