```
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import warnings
import gsw

#=================================================
# Modify paths to point to output files
#=================================================
# Case name (Straight Coast)
case_straight = 'Straight Coast'

# path to ocean_static file
path_static_straight = '/data/sragen/aquaplanet/MASTERS/straight_coast/run/DATA/1000yr/28000101.ocean_static.nc'
d_straight = xr.open_dataset(path_static_straight, decode_times=True)


# Case name (Both Coast)
case_both = 'Both Coast'

# path to ocean_static file
path_static_both = '/data/sragen/aquaplanet/MASTERS/both_coast/run/DATA/1000yr/28000101.ocean_static.nc'
d_both = xr.open_dataset(path_static_both, decode_times=True)



# Case name (Wide Coast)
case_wide = 'Wide Coast'

# path to ocean_static file
path_static_wide = '/data/sragen/aquaplanet/MASTERS/wide_straight/run/DATA/1000yr/28000101.ocean_static.nc'
d_wide = xr.open_dataset(path_static_wide, decode_times=True)



# Case name (America Coast)
case_am = 'America Coast'

# path to ocean_static file
path_static_am = '/data/sragen/aquaplanet/MASTERS/am_coast/run/DATA/900yr/27000101.ocean_static.nc'
d_am = xr.open_dataset(path_static_am, decode_times=True)



# Case name (Africa Coast)
case_af = 'Africa Coast'

# path to ocean_static file
path_static_af = '/data/sragen/aquaplanet/MASTERS/af_coast/run/DATA/1000yr/28000101.ocean_static.nc'
d_af = xr.open_dataset(path_static_af, decode_times=True)




#=================================================
# Ignore runtime warnings: mean of empty slice
#=================================================
warnings.filterwarnings("ignore", message="Mean of empty slice")

#=================================================
# Modify latitudes to point to western and eastern
# boundaries of small basin and northern extent of SO.
# Comment out for AQUA and RIDGE cases. 
#=================================================
x_west = np.where(d_both['xh']==211)[0][0]
x_east = np.where(d_both['xh']==351)[0][0]
y = np.where(d_both['yq']==-35)[0][0]
y_south = np.where(d_both['yq']==-71)[0][0]
y_north = np.where(d_both['yq']==71)[0][0]
```

```
print('Small Basin Area')

area_small_straight = (d_straight.areacello.sel(xh=slice(211,360)).sel(yh=slice(-72,72))*
                       d_straight.sftof.sel(xh=slice(211,360)).sel(yh=slice(-72,72))).sum(dim='xh').sum(dim='yh')
a_small_straight = float(area_small_straight)

print('{0:.6E}'.format(a_small_straight)+' m\u00b2')


area_small_wide = (d_wide.areacello.sel(xh=slice(211,360)).sel(yh=slice(-72,72))*
                       d_wide.sftof.sel(xh=slice(211,360)).sel(yh=slice(-72,72))).sum(dim='xh').sum(dim='yh')
a_small_wide = float(area_small_wide)

print('{0:.6E}'.format(a_small_wide)+' m\u00b2')


area_small_both = (d_both.areacello.sel(xh=slice(211,360)).sel(yh=slice(-72,72))*
                       d_both.sftof.sel(xh=slice(211,360)).sel(yh=slice(-72,72))).sum(dim='xh').sum(dim='yh')
a_small_both = float(area_small_both)

print('{0:.6E}'.format(a_small_both)+' m\u00b2')


area_small_am = (d_am.areacello.sel(xh=slice(211,360)).sel(yh=slice(-72,72))*
                       d_am.sftof.sel(xh=slice(211,360)).sel(yh=slice(-72,72))).sum(dim='xh').sum(dim='yh')
a_small_am = float(area_small_am)

print('{0:.6E}'.format(a_small_am)+' m\u00b2')


area_small_af = (d_af.areacello.sel(xh=slice(211,360)).sel(yh=slice(-72,72))*
                       d_af.sftof.sel(xh=slice(211,360)).sel(yh=slice(-72,72))).sum(dim='xh').sum(dim='yh')
a_small_af = float(area_small_af)

print('{0:.6E}'.format(a_small_af)+' m\u00b2')


##################################################################
print(' ')

print('Small Basin Area North of 35S')

area_small_straight = (d_straight.areacello.sel(xh=slice(211,360)).sel(yh=slice(-35,72))*
                       d_straight.sftof.sel(xh=slice(211,360)).sel(yh=slice(-35,72))).sum(dim='xh').sum(dim='yh')
a_small_straight = float(area_small_straight)

print('{0:.6E}'.format(a_small_straight)+' m\u00b2')


area_small_wide = (d_wide.areacello.sel(xh=slice(211,360)).sel(yh=slice(-35,72))*
                       d_wide.sftof.sel(xh=slice(211,360)).sel(yh=slice(-35,72))).sum(dim='xh').sum(dim='yh')
a_small_wide = float(area_small_wide)

print('{0:.6E}'.format(a_small_wide)+' m\u00b2')


area_small_both = (d_both.areacello.sel(xh=slice(211,360)).sel(yh=slice(-35,72))*
                       d_both.sftof.sel(xh=slice(211,360)).sel(yh=slice(-35,72))).sum(dim='xh').sum(dim='yh')
a_small_both = float(area_small_both)

print('{0:.6E}'.format(a_small_both)+' m\u00b2')


area_small_am = (d_am.areacello.sel(xh=slice(211,360)).sel(yh=slice(-35,72))*
                       d_am.sftof.sel(xh=slice(211,360)).sel(yh=slice(-35,72))).sum(dim='xh').sum(dim='yh')
a_small_am = float(area_small_am)

print('{0:.6E}'.format(a_small_am)+' m\u00b2')


area_small_af = (d_af.areacello.sel(xh=slice(211,360)).sel(yh=slice(-35,72))*
                       d_af.sftof.sel(xh=slice(211,360)).sel(yh=slice(-35,72))).sum(dim='xh').sum(dim='yh')
a_small_af = float(area_small_af)

print('{0:.6E}'.format(a_small_af)+' m\u00b2')
```

```
print('Small Basin Average Width')

width_small_straight = (d_straight.dxt*
                        d_straight.sftof).sel(xh=slice(211,360)).sel(yh=slice(-35,72)).sum(dim='xh').mean(dim='yh')
w_small_straight = float(width_small_straight)

print('{0:.6E} m'.format(w_small_straight))


width_small_wide = (d_wide.dxt*
                        d_wide.sftof).sel(xh=slice(211,360)).sel(yh=slice(-35,72)).sum(dim='xh').mean(dim='yh')
w_small_wide = float(width_small_wide)

print('{0:.6E} m'.format(w_small_wide))


width_small_both = (d_both.dxt*
                        d_both.sftof).sel(xh=slice(211,360)).sel(yh=slice(-35,72)).sum(dim='xh').mean(dim='yh')
w_small_both = float(width_small_both)

print('{0:.6E} m'.format(w_small_both))


width_small_am = (d_am.dxt*
                        d_am.sftof).sel(xh=slice(211,360)).sel(yh=slice(-35,72)).sum(dim='xh').mean(dim='yh')
w_small_am = float(width_small_am)

print('{0:.6E} m'.format(w_small_am))


width_small_af = (d_af.dxt*
                        d_af.sftof).sel(xh=slice(211,360)).sel(yh=slice(-35,72)).sum(dim='xh').mean(dim='yh')
w_small_af = float(width_small_af/1000)

print('{0:.6E} km'.format(w_small_af))

#################################################################
print(' ')
print('Small Basin Width at 60N')

width_small_straight = (d_straight.dxt*
                        d_straight.sftof).sel(xh=slice(211,360)).sel(yh=60).sum(dim='xh')
w_small_straight = float(width_small_straight)

print('{0:.6E} m'.format(w_small_straight))


width_small_wide = (d_wide.dxt*
                        d_wide.sftof).sel(xh=slice(211,360)).sel(yh=60).sum(dim='xh')
w_small_wide = float(width_small_wide)

print('{0:.6E} m'.format(w_small_wide))


width_small_both = (d_both.dxt*
                        d_both.sftof).sel(xh=slice(211,360)).sel(yh=60).sum(dim='xh')
w_small_both = float(width_small_both)

print('{0:.6E} m'.format(w_small_both))


width_small_am = (d_am.dxt*
                        d_am.sftof).sel(xh=slice(211,360)).sel(yh=60).sum(dim='xh')
w_small_am = float(width_small_am)

print('{0:.6E} m'.format(w_small_am))


width_small_af = (d_af.dxt*
                        d_af.sftof).sel(xh=slice(211,360)).sel(yh=60).sum(dim='xh')
w_small_af = float(width_small_af/1000)

print('{0:.6E} km'.format(w_small_af))
```
